```python
def create_performance_pie(df):
    
    total_questions = df['total_questions'].sum()
    correct = df['correct_answers'].sum()
    incorrect = df['incorrect_answers'].sum()
    unattempted = total_questions - (correct + incorrect)
    
    sizes = [correct, incorrect, unattempted]
    labels = ['Correct', 'Incorrect', 'Unattempted']
    colors = ['#2ecc71', '#e74c3c', '#3498db']
    explode = (0.1, 0, 0)  
    
    plt.figure(figsize=(10, 8))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=90)
    
    plt.title('Total Performance Analysis', pad=20, size=14)
    
    plt.legend(labels, title="Categories", 
              loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    plt.axis('equal')
    
    return plt

fig = create_performance_pie(user_data)
plt.show()
```

```python
def analyze_topic_strength(df):
    # Group by topic and calculate metrics
    topic_metrics = df.groupby('quiz.topic').agg({
        'accuracy': lambda x: np.mean([float(str(i).strip(' %')) for i in x]),
        'score': ['mean', 'count', 'std'],
        'correct_answers': 'sum',
        'total_questions': 'sum'
    })
    
    topic_metrics.columns = ['avg_accuracy', 'avg_score', 'attempt_count', 'score_std', 
                           'total_correct', 'total_questions']
    

    topic_metrics['success_rate'] = (topic_metrics['total_correct'] / 
                                   topic_metrics['total_questions'] * 100).round(2)
    
    # Classify topics
    def get_topic_status(row):
        if row['avg_accuracy'] >= 80 and row['score_std'] < 30:
            return 'Strong'
        elif row['avg_accuracy'] >= 60:
            return 'Moderate'
        else:
            return 'Weak'
    
    topic_metrics['status'] = topic_metrics.apply(get_topic_status, axis=1)
    
    return topic_metrics[['avg_accuracy', 'success_rate', 'attempt_count', 'status']]

topic_strength = analyze_topic_strength(performance)
print("\nTopic Strength Analysis:")
print(topic_strength.sort_values('avg_accuracy', ascending=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def create_dashboard(df):
    fig = plt.figure(figsize=(15, 10))
    
    # 1Pie Chart (top left)
    plt.subplot(2, 2, 1)
    accuracy_data = df['accuracy'].str.rstrip(' %').astype(float)
    performance_data = [
        len(accuracy_data[accuracy_data >= 80]),  # Strong
        len(accuracy_data[(accuracy_data >= 60) & (accuracy_data < 80)]),  # Moderate
        len(accuracy_data[accuracy_data < 60])  # Weak
    ]
    plt.pie(performance_data, 
            labels=['Strong', 'Moderate', 'Weak'],
            colors=['#2ecc71', '#f1c40f', '#e74c3c'],
            autopct='%1.1f%%',
            explode=(0.1, 0, 0))
    plt.title('Topic Performance Distribution')
    
    # 2.Score Distribution (top right)
    plt.subplot(2, 2, 2)
    plt.bar(range(len(df)), df['score'], color='#3498db')
    plt.title('Score Distribution')
    plt.xlabel('Quiz Attempt')
    plt.ylabel('Score')
    
    # 3.Topic-wise Accuracy (bottom left)
    plt.subplot(2, 2, 3)
    topic_accuracy = df.groupby('quiz.topic')['accuracy_value'].mean()
    topic_accuracy.plot(kind='bar', color='#9b59b6')
    plt.title('Topic-wise Average Accuracy')
    plt.xticks(rotation=45, ha='right')
    
    # 4.Progress Timeline (bottom right)
    plt.subplot(2, 2, 4)
    plt.plot(range(len(df)), df['accuracy_value'], marker='o', color='#e67e22')
    plt.title('Accuracy Progress Timeline')
    plt.xlabel('Quiz Attempt')
    plt.ylabel('Accuracy (%)')
    
    plt.tight_layout()
    return plt

user_data['accuracy_value'] = user_data['accuracy'].str.rstrip(' %').astype(float)

dashboard = create_dashboard(user_data)
plt.show()
```

```python
def generate_student_insights_weak(df):
    df['accuracy_value'] = df['accuracy'].str.rstrip(' %').astype(float)
    
    # Topic-wise metrics
    topics = df['quiz.topic'].unique()
    topic_metrics = {}
    
    for topic in topics:
        topic_data = df[df['quiz.topic'] == topic]
        topic_metrics[topic] = {
            'avg_accuracy': topic_data['accuracy_value'].mean(),
            'avg_speed': topic_data['speed'].astype('int64').mean(),
            'avg_score': topic_data['score'].mean(),
            'score_std': topic_data['score'].std(),
            'total_correct': topic_data['correct_answers'].sum(),
            'total_incorrect': topic_data['incorrect_answers'].sum(),
            'total_questions': topic_data['total_questions'].sum()
        }
    
    avg_speed = df['speed'].astype('int64').mean()
    avg_accuracy = df['accuracy_value'].mean()
    score_std = df['score'].std()
    
    df['sequential_index'] = range(len(df))
    
    persona = {
        'learning_style': 'Speed-focused' if avg_speed > 95 else 'Accuracy-focused',
        'consistency': 'High' if score_std < 20 else 'Variable',
        'pattern': 'Improving' if df['score'].corr(df['sequential_index']) > 0 else 'Needs attention'
    }
    
    recommendations = []
    strength_labels = {}
    weak_topics = []
    improve_topics = []
    
    for topic, metrics in topic_metrics.items():
        if metrics['avg_accuracy'] < 60:
            weak_topics.append(topic)
            recommendations.append({
                'topic': topic,
                'priority': 'High',
                'suggestion': f"Master basics in {topic}",
                'strategy': 'Start with fundamentals'
            })
        elif metrics['avg_accuracy'] < 80:
            improve_topics.append(topic)
            recommendations.append({
                'topic': topic,
                'priority': 'Medium',
                'suggestion': f"Practice {topic}",
                'strategy': 'Focus on weak areas'
            })
        else:
            strength_labels[topic] = " Master" if metrics['avg_accuracy'] > 90 else "Proficient"
    
    return {
        'persona': persona,
        'recommendations': recommendations,
        'strength_labels': strength_labels,
        'weak_topics': weak_topics,
        'improve_topics': improve_topics
    }

insights = generate_student_insights_weak(user_data)

print("Student Persona:")
for key, value in insights['persona'].items():
    print(f"{key}: {value}")

print("\nRecommendations:")
for rec in insights['recommendations']:
    print(f"\nTopic: {rec['topic']}")
    print(f"Priority: {rec['priority']}")
    print(f"Suggestion: {rec['suggestion']}")
    print(f"Strategy: {rec['strategy']}")

print("\nStrengths:")
for topic, label in insights['strength_labels'].items():
    print(f"{label} in {topic}")

print("\nWeak Topics:")
for topic in insights['weak_topics']:
    print(f"Weak in {topic}")

print("\nTopics to Improve:")
for topic in insights['improve_topics']:
    print(f"Needs Improvement in {topic}")
```

```python
topic_stats = user_data.groupby("quiz.topic").agg({
    "correct_answers": "sum",
    "incorrect_answers": "sum",
    "total_questions": "sum"
})

topic_stats["attempted_questions"] = topic_stats["correct_answers"] + topic_stats["incorrect_answers"]
topic_stats["accuracy"] = (topic_stats["correct_answers"] / topic_stats["attempted_questions"]) * 100
topic_stats["attempt_percentage"] = (topic_stats["attempted_questions"] / topic_stats["total_questions"]) * 100

def classify_topic(row):
    if row["accuracy"] >= 80 and row["attempt_percentage"] >= 20:
        return "Strong"
    else:
        return "Weak"

topic_stats["classification"] = topic_stats.apply(classify_topic, axis=1)

detailed_insights_df = topic_stats.reset_index()[["quiz.topic", "accuracy", "attempt_percentage", "classification"]]
detailed_insights_df.columns = ["Topic", "Accuracy (%)", "Attempt Percentage (%)", "Classification"]

detailed_insights = topic_stats[["accuracy", "attempt_percentage", "classification"]].to_dict(orient="records")

print("Detailed Insights for Dashboard:")
detailed_insights
```
